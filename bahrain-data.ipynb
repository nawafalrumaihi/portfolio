```python
import pandas as pd
import plotly.express as px

# Data preprocessing
# Read the CSV file with the correct encoding
data_df = pd.read_csv('marine 2007 - 2016 may 2017.csv', encoding='ISO-8859-1')

# Convert 'Date' to datetime
data_df['Date'] = pd.to_datetime(data_df['Date'], errors='coerce')

# Replace commas and convert 'Salinity %0' to numeric
data_df['Salinity %0'] = pd.to_numeric(data_df['Salinity %0'].str.replace(',', ''), errors='coerce')

# Drop rows with missing or invalid 'Date' or 'Salinity %0'
data_df = data_df.dropna(subset=['Date', 'Salinity %0'])

# average salinity
avg_salinity_bahrain = data_df['Salinity %0'].mean()

# average salinity on ocean
avg_salinity_ocean = 35

# calculate the percent difference betweeen avg_salinity_bahrain and avg_salinity_ocean
percent_difference = int((avg_salinity_bahrain - avg_salinity_ocean) / avg_salinity_bahrain * 100)

# plot avg_salinity in the trendline plot
fig = px.scatter(data_df, x='Date', y='Salinity %0', trendline='ols', trendline_scope='overall', title='Salinity Linear Trendline')
fig.show()

print('Percent difference between average salinity in Bahrain and average salinity of the entire oceans: {:.2f}%'.format(percent_difference))
```
